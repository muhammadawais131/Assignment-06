**Task 1: Build a Basic RESTful API in Flask**

Objective: Create a Flask API with GET and POST endpoints to manage a to-do list stored in memory.

Requirements:

*  GET /todos: Return a list of all to-do items.
*  POST /todos: Add a new to-do item with a task field in the request body.

*   Store to-do items in a list (e.g., [{"id": 1, "task": "Buy groceries"}]) in memory.
*   Return JSON responses with appropriate status codes (e.g., 200 OK, 400 Bad Request).

*   Test using curl or a browser.

**Solution :**  **CODE :**



```python
from flask import Flask, request, jsonify

app = Flask(__name__)

# In-memory storage for todos
todos = []
current_id = 1

@app.route('/todos', methods=['GET'])
def get_todos():
    """GET endpoint to return all todos"""
    return jsonify(todos), 200

@app.route('/todos', methods=['POST'])
def add_todo():
    """POST endpoint to add a new todo"""
    global current_id


    if not request.is_json:
        return jsonify({"error": "Request must be JSON"}), 400

    data = request.get_json()


    if 'task' not in data or not data['task']:
        return jsonify({"error": "Task field is required"}), 400

    # Create new todo
    new_todo = {
        "id": current_id,
        "task": data['task']
    }

    todos.append(new_todo)
    current_id += 1

    return jsonify(new_todo), 201

if __name__ == '__main__':
    app.run(debug=True, port=5000)
```

**Task 2: Rewrite Task 1 in FastAPI with Type Hints**

Objective: Rewrite the Flask API from Task 1 using FastAPI, adding type hints and testing via auto-generated docs.

Requirements:

GET /todos: Return the list of to-do items.

POST /todos: Add a new to-do with a task field, using a Pydantic model for validation.

Use type hints and Pydantic for data validation.

Store to-dos in memory (same structure as Task 1).

Test using FastAPIâ€™s Swagger UI at /docs.

**CODE : **

```python
from fastapi import FastAPI, HTTPException
from pydantic import BaseModel
from typing import List, Optional

app = FastAPI(title="Todo API", description="FastAPI Todo App with Type Hints")


class TodoItem(BaseModel):
    task: str

class TodoResponse(BaseModel):
    id: int
    task: str

# In-memory storage
todos: List[TodoResponse] = []
current_id = 1

@app.get("/todos", response_model=List[TodoResponse])
def get_todos():
    """GET endpoint to return all todos"""
    return todos

@app.post("/todos", response_model=TodoResponse, status_code=201)
def add_todo(todo: TodoItem):
    """POST endpoint to add a new todo"""
    global current_id

    # new todo
    new_todo = TodoResponse(
        id=current_id,
        task=todo.task
    )

    todos.append(new_todo)
    current_id += 1

    return new_todo

if __name__ == "__main__":
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
```

Task 3: Extend Either API with Authentication and DELETE Endpoint

Objective: Extend the Flask or FastAPI to-do API with a simple API key authentication and a DELETE endpoint.

Requirements:

Add a simple API key check (e.g., check for a header X-API-Key: secret).

DELETE /todos/{id}: Remove a to-do item by its ID.

Return appropriate status codes (e.g., 404 if ID not found, 401 for invalid API key).

Test using curl or Postman.

**Task 03 SOLUTION :**

**CODE FOR API :**

```python
from fastapi import FastAPI, HTTPException, Header, Depends
from pydantic import BaseModel
from typing import List

app = FastAPI(title="Todo API with Auth", description="FastAPI Todo App with Authentication")


class TodoItem(BaseModel):
    task: str

class TodoResponse(BaseModel):
    id: int
    task: str


todos: List[TodoResponse] = []
current_id = 1

# API Key configuration
API_KEY = "secret123"

def verify_api_key(api_key: str = Header(..., alias="X-API-Key")):
    """Dependency to verify API key"""
    if api_key != API_KEY:
        raise HTTPException(status_code=401, detail="Invalid API Key")
    return api_key

@app.get("/todos", response_model=List[TodoResponse])
def get_todos():
    """GET endpoint to return all todos (No auth required)"""
    return todos

@app.post("/todos", response_model=TodoResponse, status_code=201)
def add_todo(todo: TodoItem, api_key: str = Depends(verify_api_key)):
    """POST endpoint to add a new todo (Auth required)"""
    global current_id

    new_todo = TodoResponse(
        id=current_id,
        task=todo.task
    )

    todos.append(new_todo)
    current_id += 1

    return new_todo

@app.delete("/todos/{todo_id}")
def delete_todo(todo_id: int, api_key: str = Depends(verify_api_key)):
    """DELETE endpoint to remove a todo by ID (Auth required)"""
    global todos

    # find to do by ID
    for index, todo in enumerate(todos):
        if todo.id == todo_id:
            deleted_todo = todos.pop(index)
            return {"message": f"Todo with ID {todo_id} deleted", "deleted_todo": deleted_todo}

    # If todo not found
    raise HTTPException(status_code=404, detail=f"Todo with ID {todo_id} not found")

if __name__ == "__main__":
    import uvicorn
    uvicorn.run(app, host="0.0.0.0", port=8000)
```
